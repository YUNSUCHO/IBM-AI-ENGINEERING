```python
import torch
import torch.nn as nn
from torch import sigmoid
import matplotlib.pyplot as plt
import numpy as np
torch.manual_seed(0)
```

```python
# The function for plotting the model

def PlotStuff(X, Y, model, epoch, leg = True):
    plt.plot(X.numpy(), model(X).detach().numpy(), label = ('epoch' +str(epoch)))
    plt.plot(X.numpy(), Y.numpy(), 'r')
    plt.xlabel('x')
    if leg == True:
        plt.legend()
    else:
        pass
```

```python
#NeuralNetwork Module and training function
```

```python
# Define the class Net

class Net(nn.Module):
    
    #constructor
    def __init__(self, D_in, H, D_out):
        super(Net, self).__init__()
        #hidden layer
        self.linear1 = nn.Linear(D_in, H)
        self.linear2 = nn.Linear(H, D_out)
        #Define the first linear layer as an attribute
        self.a1 = None
        self.l1 = None
        self.l2 = None
        
    #prediction
    def forward(self, x):
        self.l1 = self.linear1(x)
        self.a1 = sigmoid(self.l1)
        self.l2 = self.linear2(self.a1)
        
        yhat = sigmoid(self.linear2(self.a1))
        return yhat
```

```python
# Define the training function

def train(X, Y, model, optimizer, criterion, epochs = 1000):
    cost = []
    total = 0
    for epoch in range(epochs):
        total = 0
        for x, y in zip(X, Y):
            yhat = model(x)
            loss = criterion(yhat, y)
            loss.backward()
            optimizer.step()
            optimizer.zero_grad()
            
            total+= loss.item()
        cost.append(total)
        if epoch % 300 == 0:
            PlotStuff(X, Y, model, epoch, leg = True)
            plt.show()
            model(X)
            plt.scatter(model.a1.detach().numpy()[:, 0], model.a1.detach().numpy()[:, 1], c=Y.numpy().reshape(-1))
            plt.title('activations')
            plt.show()
    return cost
```

```python
# Make some data

X = torch.arange(-20, 20, 1).view(-1, 1).type(torch.FloatTensor)
Y = torch.zeros(X.shape[0])
Y[(X[:, 0] > -4) & (X[:, 0] < 4)] = 1.0
```

```python
# The loss function
#Create cross-entropy loss

def criterion_cross(outputs, labels):
    out = -1 * torch.mean(labels * torch.log(outputs) + (1 - labels) * torch.log(1 - outputs))
    return out
```

```python
# Train the model

D_in = 1
H = 2
D_out = 1
learning_rate = 0.1

model = Net(D_in, H, D_out)
optimizer = torch.optim.SGD(model.parameters(), lr = learning_rate)
cost_cross = train(X, Y, model, optimizer, criterion_cross, epochs = 1000)

plt.plot(cost_cross)
plt.xlabel('epochs')
plt.title('cross entropy loss')
```
